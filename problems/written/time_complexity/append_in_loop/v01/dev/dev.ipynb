```python
import time
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bar(n):
    lst = []
    for i in range(n):
        lst.append(i**2)
    return np.array(lst)

def time_taken(n):
    start = time.time()
    bar(n)
    stop = time.time()
    return stop - start

sizes = [10_000, 20_000, 40_000, 80_000, 120_000, 160_000]
times = [time_taken(n) for n in sizes]

plt.plot(sizes, times)
plt.xlabel("n")
plt.ylabel("seconds")
plt.savefig("times-bar.pdf")
```
