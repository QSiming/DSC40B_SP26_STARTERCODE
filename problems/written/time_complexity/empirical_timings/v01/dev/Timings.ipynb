```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def find_median(numbers):
    n = len(numbers)
    for x in numbers:
        less = 0
        more = 0
        for y in numbers:
            if y >= x:
                more += 1
            if y <= x:
                less += 1
        if less >= n//2 and more >= n//2:
            return x
```

```python
def time_median(data):
    start = time.time()
    find_median(data)
    return time.time() - start
```

```python
def average_time_on_n(n, trials=20):
    times = []
    for _ in range(trials):
        data = np.random.randint(0, 10_000, n)
        t = time_median(data)
        times.append(t)
    return sum(times) / trials

sizes = range(100, 2001, 100)
times = [average_time_on_n(s) for s in sizes]

plt.plot(sizes, times)
plt.xlabel('n')
plt.ylabel('time')
plt.savefig('../include-solution/timing.pdf', bbox_inches='tight')
```
